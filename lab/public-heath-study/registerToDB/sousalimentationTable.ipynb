```python
##sous ailementaation traitement


import pandas as pd
import numpy as np
import pymysql as pymysql
import mysql.connector
from mysql.connector import Error



sousAlimentationCsv = pd.read_csv("../data/fr_sousalimentation.csv")
sousAlimentationCsv=sousAlimentationCsv[['Code année','Code zone','Zone','Produit','Valeur','Année']]
sousAlimentationCsv["Valeur"] = pd.to_numeric(sousAlimentationCsv["Valeur"],errors='coerce')
sousAlimentationCsv['underfeed'] = sousAlimentationCsv['Valeur'] * 100000
sousAlimentationCsv = sousAlimentationCsv.loc[(sousAlimentationCsv['Code année'] == 20122014 )]
sousAlimentationCsv = sousAlimentationCsv.groupby(['Zone']).sum()
sousAlimentationCsv= sousAlimentationCsv.rename(columns={"Zone":"pays","Code zone":"code_pays","Année":"annee","underfeed":"nb_personnes"})
sousAlimentationCsv
```

```python
df_population = pd.read_csv("../data/fr_population.csv")
population = df_population[['Zone','Valeur','Code zone','Année']]
population = population[population['Valeur'] > 0]
population['Valeur']= population['Valeur']*1000
population=population.rename(columns={"Zone":"pays","Code zone":"code_pays","Année":"annee","Valeur":"population"})
population.drop_duplicates(subset=['pays'], keep= False)
population=population.loc[population['pays']!= 'Chine',:]
population =  population[['pays','code_pays','annee','population']]

population
```

```python
merge = sousAlimentationCsv.merge(population,on="code_pays")

merge
```

```python
sousAlim = merge[['pays','code_pays','annee','nb_personnes']]


sousAlim
```

```python
try:
    connect = mysql.connector.connect(   host='db',
                                         database='health',
                                         user='root',
                                         password='root'
                                  )
    if connect.is_connected():
        cursor = connect.cursor()
        cols = "`,`".join([str(i) for i in sousAlim.columns.tolist()])
    
        for i,ssAlim in sousAlim.iterrows():

            sql = "INSERT INTO `sous_nutrition` (`" +cols + "`) VALUES (" + "%s,"*(len(ssAlim)-1) + "%s)"
            cursor.execute(sql, tuple(ssAlim))
            connect.commit()
                
            
except Error as e:
    print("Error while connecting to MySQL", e)
finally:
    if (connect.is_connected()):
        cursor.close()
        connect.close()
        print("MySQL connection is closed")
```
